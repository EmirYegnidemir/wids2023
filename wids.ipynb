```python
import pandas as pd
train = pd.read_csv("train_data.csv")
```

```python
test = pd.read_csv("test_data.csv")
```

```python
from datetime import datetime

def add_date_feats(data):

    data["month"]=pd.to_datetime(data["startdate"]).dt.month
    data["week"]=pd.to_datetime(data["startdate"]).dt.week
    data["year"]=pd.to_datetime(data["startdate"]).dt.year
    data["day"]=pd.to_datetime(data["startdate"]).dt.day
    data["day_of_weeeks"]=pd.to_datetime(data["startdate"]).dt.day_of_week
    data["quarter"]=pd.to_datetime(data["startdate"]).dt.quarter
    data["hour"]=pd.to_datetime(data["startdate"]).dt.hour
    #data = data.drop(['startdate',"climateregions__climateregion"],axis=1)
    return data

train = add_date_feats(train)
test = add_date_feats(test)
```

```python
# Concatenate train and test data
all_df = pd.concat([train, test], axis=0)

# Create new feature
all_df['loc_group'] = all_df.groupby(['lat','lon']).ngroup()
display(all_df)

# Split back up
train = all_df.iloc[:len(train)]
test = all_df.iloc[len(train):]

del all_df
```

```python
train['nmme0-prate-34w__ccsm30'] = train['nmme0-prate-34w__ccsm30'].fillna(train['nmme0-prate-34w__ccsm30'].mean())
train['nmme0-tmp2m-34w__ccsm30'] = train['nmme0-tmp2m-34w__ccsm30'].fillna(train['nmme0-tmp2m-34w__ccsm30'].mean())
train['ccsm30'] = train['ccsm30'].fillna(train['ccsm30'].mean())
train['nmme0-prate-56w__ccsm30'] = train['nmme0-prate-56w__ccsm30'].fillna(train['nmme0-prate-56w__ccsm30'].mean())
train['nmme-tmp2m-56w__ccsm3'] = train['nmme-tmp2m-56w__ccsm3'].fillna(train['nmme-tmp2m-56w__ccsm3'].mean())
train['nmme-prate-56w__ccsm3'] = train['nmme-prate-56w__ccsm3'].fillna(train['nmme-prate-56w__ccsm3'].mean())
train['nmme-tmp2m-34w__ccsm3'] = train['nmme-tmp2m-34w__ccsm3'].fillna(train['nmme-tmp2m-34w__ccsm3'].mean())
train['nmme-prate-34w__ccsm3'] = train['nmme-prate-34w__ccsm3'].fillna(train['nmme-prate-34w__ccsm3'].mean())
```

```python
#converting object type to category
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
train['climateregions__climateregion'] = le.fit_transform(train['climateregions__climateregion'])
test['climateregions__climateregion'] = le.transform(test['climateregions__climateregion'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import numpy as np
import matplotlib as plt

num_climate_groups = train["climateregions__climateregion"].nunique()
print("Number of unique climate groups:", num_climate_groups)
target="contest-tmp2m-14d__tmp2m"
exclude_cols = ['index', 'startdate', 'lat', 'lon']
features = [c for c in train.columns if ((c != target) & (c not in exclude_cols))]
#model definition
for climate_group in range(num_climate_groups):
    scaler = StandardScaler()
    filtered_train = train[train['climateregions__climateregion'] == climate_group]
    filtered_test = test[test['climateregions__climateregion'] == climate_group]
    X_train, X_test, y_train, y_test = train_test_split(filtered_train[features], filtered_train[target],
                                                    test_size=0.33, random_state=42)
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                               n_estimators=3000,
                               #early_stopping_rounds=25,
                               objective='reg:squarederror',
                               tree_method = 'auto',
                               max_depth=3,
                               #gpu_id=0,
                               learning_rate=0.1)
    reg.fit(X_train, y_train, verbose=100,  eval_set=[(X_train, y_train), (X_test, y_test)])
    y_pred_val = reg.predict(X_test)
    score = np.sqrt(mean_squared_error(y_test, y_pred_val))
    print(f'RMSE Score on Test set for the region {climate_group}: {score:0.2f}')
    # Save the results
    np.save(f"result_region_{climate_group}.npy", y_pred_val)
    reg.save_model("model"+str(climate_group)+".txt")
    thresh = 0
list_vals = reg.feature_importances_[reg.feature_importances_>=thresh]
list_vals
prediction_results =[]
for climate_group in range(num_climate_groups):
    filtered_test = test[test['climateregions__climateregion'] == climate_group]
    prediction = reg.predict(filtered_test[features])
    # Print the results for each climate group
    print(f'Results for climate group {climate_group}:')
    # Save the results for each climate group to a separate csv file
    for i in range(len(prediction)):
        prediction_results.append([filtered_test.index[i], climate_group, prediction[i]])

df = pd.DataFrame(prediction_results, columns=["index", "climate_group", "prediction"])
df.to_csv("prediction_results.csv", index=False)
```

```python
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
target="contest-tmp2m-14d__tmp2m"
exclude_cols = ['index', 'startdate', 'lat', 'lon']
features = [c for c in train.columns if ((c != target) & (c not in exclude_cols))]
scalers = {}
regressors = {}
num_climate_groups = train["climateregions__climateregion"].nunique()
prediction_results = []
submission = pd.read_csv('sample_solution.csv')

for climate_group in range(num_climate_groups):
    filtered_train = train[train['climateregions__climateregion'] == climate_group]
    filtered_test = test[test['climateregions__climateregion'] == climate_group]

    # Load the scaler
    scaler = StandardScaler()
    scalers[climate_group] = scaler
    scaler.fit(filtered_train[features])
    X_test_scaled = scaler.transform(filtered_test[features])

    # Load the regressor
    reg = xgb.XGBRegressor()
    regressors[climate_group] = reg
    reg.load_model(f"model{climate_group}.txt")
    # Make predictions using the loaded model and transformed test data
    prediction = reg.predict(X_test_scaled)
    submission['contest-tmp2m-14d__tmp2m'][filtered_test.index] = prediction

# Write the DataFrame to a csv file
submission.to_csv('submission2.csv',
                  index = False) # Set index to false to avoid issues in evaluation
```
